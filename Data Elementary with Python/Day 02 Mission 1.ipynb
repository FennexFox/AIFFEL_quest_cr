```python
import seaborn as sns

data = sns.load_dataset('taxis')
```

```python
data.info()
```

```python
na_series = data.isna().sum()
print(na_series.max() / 6433)
```

```python
na_df_indices = {}

for i, v in zip(na_series.index, na_series):
    if v == 0:
        na_series.drop(i, inplace = True)
    else:
        print(f'N/A of {i}: {v/6433*100:.2f}% of {data.index.size} ent, total of {v}')
        na_df_indices.update({i: data[data[i].isna()].index})

data.loc[na_df_indices['pickup_zone']]
```
