```python
import math
import random

class Character:
  def __init__(self, name: str, level, health, attack, defence):
    self.name = name
    self.level = level
    self.health = health
    self.attack = attack
    self.defence = defence

  def is_alive(self):
    return (lambda x: True if x >= 0 else False)(self.health)

  def take_damage(self, damage):
    damage_point = max(damage - self.defence, 0)

    print(f'{self.name}(이/가) {damage_point}만큼 피해를 입었다!')
    self.health -= damage_point
    print(f'{self.name}의 체력: {self.health}')

  def attack_target(self, target):
    try:
      attack_point = random.randint(1, self.attack)
      print(f'{self.name}(이/가) {attack_point}만큼 공격했다!')
      target.take_damage(attack_point)
    except Exception as e:
      print(f'{target} is not a valid target: {e}')

class Player(Character):
  def __init__(self, name: str):
    super().__init__(name, 1, 100, 25, 5)
    self.exp = 0

  def gain_experience(self, exp: int):
    try:
      exp = int(exp)
      if exp < 1:
        raise Exception('경험치가 양수가 아닙니다.')
      else:
        self.exp += exp
        self.level_up()
    except Exception as e:
      print(f'잘못된 경험치 입력입니다: {e}')

  def level_up(self):
    if self.exp >= 50:
      self.level += 1
      self.attack += 10
      self.defence += 5
      # self.health = 100
      self.exp = 0

      print("레벨 업!")

class Monster(Character):
  def __init__(self, name: str, level: int):
    health = random.randint(10, 30) * level
    attack = random.randint(5, 20) * level
    defence = random.randint(1, 5) * level
    super().__init__(name, level, health, attack, defence)
```

```python
def battle(player, monster):
  print(f'\n{monster.name} 상대로 전투를 시작합니다.')

  try:
    while player.is_alive() and monster.is_alive():
      player.attack_target(monster)
      monster.attack_target(player)
  except Exception as e:
    print(f'유효한 전투가 아닙니다: {e}')

  if player.is_alive():
    player.gain_experience(monster.level * 20)
    print("전투 승리!")
  else:
    print("전투 패배..")

def main():
  monster_dict = {'슬라임':1, '고블린':2, '오크':3}
  player = Player(str(input("주인공 캐릭터의 이름을 입력하세요:")))

  for name, level in monster_dict.items():
    monster_dict[name] = Monster(name, level)

  for name, monster in monster_dict.items():
    battle(player, monster)
    if not player.is_alive():
      print('게임오버')
      break

if __name__ == '__main__':
  main()
```
