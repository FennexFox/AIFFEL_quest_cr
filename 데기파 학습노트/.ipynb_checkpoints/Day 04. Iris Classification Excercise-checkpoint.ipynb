````python
# sklearn model_selection train_test_split  
from sklearn.model_selection import train_test_split

#   (, X): iris_data
#  (, y): iris_label
# iris_data iris_label  train:test = 8:2   
# X_train, X_test, y_train, y_test 
X_train, X_test, y_train, y_test = train_test_split(iris_data, 
                                                    iris_label, 
                                                    test_size=0.2, 
                                                    random_state=7)

# len   
print('X_train : ', len(X_train),', X_test : ', len(X_test))
````

````python
# train  
print(X_train.shape, y_train.shape)

# test  
print(X_test.shape, y_test.shape)

# label   
print(y_train, y_test, sep = "\n")
````

````python
# sklearn.tree    import
from sklearn.tree import DecisionTreeClassifier 

# random_state :    32 
decision_tree = DecisionTreeClassifier(random_state=32) 
print(decision_tree._estimator_type)
````

````python
#  X_train, y_train   
decision_tree.fit(X_train, y_train)
````

---

## 5-6.   

---

````python
#  X_test 
y_pred = decision_tree.predict(X_test) #feature set X-test  label set 
y_pred
````

````python
#  : 
#  ...

pct = 0
for pred, true in zip(y_pred, y_test):
    if pred == true:
        pct += 1

print(pct/len(y_pred))
````

````python
#  :  
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
````

````python
#   :  
from sklearn.metrics import classification_report

report = classification_report(y_test, y_pred)
print(report)
````

```python
# (1)   import
from sklearn.datasets import load_iris #   iris  
from sklearn.model_selection import train_test_split #   train test dataset   import
from sklearn.tree import DecisionTreeClassifier #       
from sklearn.metrics import classification_report #        

# (2)  
iris = load_iris() #iris   .
iris_data = iris.data #iris data  iris_data  .
iris_label = iris.target #iris target  iris_label  .

# (3) train, test  
#train_test_split()  X, y  train data test data .     x,y 
#test_size    , random_state    . (   .)
X_train, X_test, y_train, y_test = train_test_split(iris_data, 
                                                    iris_label, 
                                                    test_size=0.2, 
                                                    random_state=7) 


# (4)    
decision_tree = DecisionTreeClassifier(random_state=32) #  .
decision_tree.fit(X_train, y_train) #  x y     .
y_pred = decision_tree.predict(X_test) #   X_test      .

print(classification_report(y_test, y_pred)) #     classification_report  y_test, y_pred   .
```

### 5-7-1. (Random Forest)
![rf_example](https://1.bp.blogspot.com/-Ax59WK4DE8w/YK6o9bt_9jI/AAAAAAAAEQA/9KbBf9cdL6kOFkJnU39aUn4m8ydThPenwCLcBGAsYHQ/s0/Random%2BForest%2B03.gif)

-  :           ,  : 

---

````python
from sklearn.ensemble import RandomForestClassifier #    import

X_train, X_test, y_train, y_test = train_test_split(iris_data, # iris  data 
                                                    iris_label, # iris  target 
                                                    test_size=0.2, # test_size : train data test data    
                                                    random_state=21) # random_state :    

random_forest = RandomForestClassifier(random_state=32) # RandomForest  
random_forest.fit(X_train, y_train) # 
y_pred = random_forest.predict(X_test) # 

print(classification_report(y_test, y_pred)) #   
````
